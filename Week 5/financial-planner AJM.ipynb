### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
# YOUR CODE HERE!
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# YOUR CODE HERE!
btc_price = requests.get(btc_url)
btc_price_data = btc_price.json()
#print(json.dumps(btc_price_data, indent=4))
btc_price = btc_price_data["data"]["1"]["quotes"]["USD"]["price"]

# Fetch current ETH price
# YOUR CODE HERE!]
eth_price = requests.get(eth_url)
eth_price_data = eth_price.json()
#print(json.dumps(eth_price_data, indent=4))
eth_price = eth_price_data["data"]["1027"]["quotes"]["USD"]["price"]

# Compute current value of my crpto
# YOUR CODE HERE!
my_btc_wallet = my_btc*btc_price
my_eth_wallet = my_eth* eth_price

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_wallet:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_wallet:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Current amount of shares
# YOUR CODE HERE!
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
# YOUR CODE HERE!
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
#print(f"Alpaca Key type: {type(alpaca_api_key)}")
#print(f"Alpaca Secret Key type: {type(alpaca_secret_key)}")

# Create the Alpaca API object
# YOUR CODE HERE!
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
# YOUR CODE HERE!
start = pd.Timestamp("2019-10-20", tz="America/New_York").isoformat()
end = pd.Timestamp("2020-10-20", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# YOUR CODE HERE!
df_portfolio = alpaca.get_barset(
    tickers,
    timeframe,
    start = start,
    end = end
).df

# Preview DataFrame
# YOUR CODE HERE!
df_portfolio
```

```python
# Pick AGG and SPY close prices
# YOUR CODE HERE!
agg_close_price = df_portfolio.iloc[-1, 3]
spy_close_price = df_portfolio.iloc[-1, 8]

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# YOUR CODE HERE!

my_spy_value = my_spy*spy_close_price
my_agg_value = my_agg*agg_close_price

# Print current value of share
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2015-08-07', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2020-08-07', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# YOUR CODE HERE!
df_stock_data = alpaca.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Display sample data
df_stock_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
num_sims = 500
MC_my_port = MCSimulation(
    portfolio_data = df_stock_data,
    weights = [.4, .6],
    num_simulation = num_sims,
    num_trading_days = 7560
)
```

```python
# Printing the simulation input data
# YOUR CODE HERE!
MC_my_port.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_my_port.calc_cumulative_return()
```

```python
# Plot simulation outcomes
# YOUR CODE HERE!
line_plot = MC_my_port.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
# YOUR CODE HERE!
dist_plot = MC_my_port.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!
even_tbl = MC_my_port.summarize_cumulative_return()

# Print summary statistics
# YOUR CODE HERE!
print(even_tbl)
```

### Calculate the expected portfolio return at the 95% lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
# YOUR CODE HERE!
conf_in_lower = round(even_tbl[8]*initial_investment,2)
conf_in_upper = round(even_tbl[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${conf_in_lower} and ${conf_in_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment2 = (20000 * 1.5)

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!
conf_in_lower2 = round(even_tbl[8]*initial_investment2,2)
conf_in_upper2 = round(even_tbl[9]*initial_investment2,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment2} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${conf_in_lower2} and ${conf_in_upper2}")
```
